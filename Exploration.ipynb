```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_hdf('Data.h5')
```

```python
data.head()
```

```python
plt.figure(figsize=(20,10))
plt.bar(range(data.shape[1]), data.mean())
```

```python
means_without_largest = np.array(data.mean())
for _ in range(4):
    means_without_largest = np.delete(means_without_largest, np.argmax(means_without_largest))
plt.figure(figsize=(20,10))
plt.bar(range(len(means_without_largest)), means_without_largest)
```

```python
plt.figure(figsize=(20,10))
plt.bar(range(data.shape[1]), data.median())
```

```python
#Большая разница между медианой и средним говорит о наличии выбросов по краям распределения
plt.figure(figsize=(20,10))
plt.bar(range(data.shape[1]), data.mean() - data.median())
```

```python
plt.figure(figsize=(20,10))
plt.bar(range(data.shape[1]), data.std())
```

```python
#Как сильно максимум отклоняется от среднего плюс 2 стандартных отклонения
plt.figure(figsize=(20,10))
plt.bar(range(data.shape[1]), data.max() - data.mean() - 2*data.std())
```

```python
#Как сильно минимум отклоняется от среднего минус 2 стандартных отклонения
plt.figure(figsize=(20,10))
plt.bar(range(data.shape[1]), data.mean() - 2 * data.std() - data.min())
```

```python
# Нормализуем данные, чтобы посмотреть относительные значения статистик
normalized_data=(data-data.min())/(data.max()-data.min())
```

```python
plt.figure(figsize=(20,10))
plt.bar(range(data.shape[1]), normalized_data.mean())
```

```python
plt.figure(figsize=(20,10))
plt.bar(range(data.shape[1]), normalized_data.mean() - normalized_data.median())
```

```python
plt.figure(figsize=(20,10))
plt.bar(range(data.shape[1]), normalized_data.std())
```

```python
plt.figure(figsize=(20,10))
plt.bar(range(data.shape[1]), normalized_data.max() - normalized_data.mean() - 2*normalized_data.std())
```

```python
plt.figure(figsize=(20,10))
plt.bar(range(data.shape[1]), normalized_data.mean() - 2*normalized_data.std() - normalized_data.min())
```

```python
rolling_means = data.rolling(100).mean()
```

```python
# Резкое изменение rolling mean происходит в момент резкого изменения основного графика, 
# но пока окно оттуда не уйдет, скачок будет влиять на среднее
plt.figure(figsize=(14,10))
i=np.random.randint(data.shape[1])
print(i)
print(data.columns[i])
plt.plot(data.index, data.iloc[:,i])
plt.plot(data.index, rolling_means.iloc[:,i], color='red')
```

```python
cumul_means = data.expanding().mean()
```

```python
# Скачки ощутимо меняют значение кумулятивного среднего,
# но чем дальше по времени, тем слабее этот эффект
plt.figure(figsize=(14,10))
i=np.random.randint(data.shape[1])
print(i)
print(data.columns[i])
plt.plot(data.index, data.iloc[:,i])
plt.plot(data.index, cumul_means.iloc[:,i], color='red')
```

```python
from statsmodels.tsa.arima_model import ARIMA
i=np.random.randint(data.shape[1])
print(i)
print(data.columns[i])
d=0
arima = ARIMA(data.iloc[:,i], order=(25,d,10))
arima = arima.fit()

plt.figure(figsize=(14,10))
plt.plot(data.index, data.iloc[:,i])
plt.plot(data.index[d:], arima.resid, color='red')
```
